```python
# List of names to write to the file
names = ["Anitha",
         "Bryan", 
         "Kevin",
         "Faela",
         "Esther"]

# Open the file "data.txt" for writing
with open("data.txt", "w") as file:
    # Write each name to the file, one name per line
    for name in names:
        file.write(name + "\n")
```

```python
# Initialize an empty list to store names
names = []

# Initialize variables to keep track of the total number of names and the longest name
total_names = 0
longest_name = ""

# Open the file "data.txt" for reading
with open("data.txt", "r") as file:
    # Loop through each line in the file
    for line in file:
        # Remove leading and trailing whitespaces (e.g., newline characters)
        name = line.strip()
        
        # Append the name to the list of names
        names.append(name)
        
        # Update the total number of names
        total_names += 1
        
        # Check if the current name is longer than the longest name encountered so far
        if len(name) > len(longest_name):
            longest_name = name

# Print the total number of names and the longest name
print("Total number of names:", total_names)
print("Longest name:", longest_name)
```

```python
def find_shortest_name(names):
    # Initialize variables to store the shortest length and a list of names with that length
    shortest_length = float('inf')
    shortest_names = []

    for name in names:
        name_length = len(name)
        if name_length < shortest_length:
            # Reset the list with the new shortest name
            shortest_names = [name]
            shortest_length = name_length
        elif name_length == shortest_length:
            # Append the name to the list of shortest names
            shortest_names.append(name)

    return shortest_names

# Initialize an empty list to store names
names = []

# Open the file "data.txt" for reading
with open("data.txt", "r") as file:
    # Loop through each line in the file
    for line in file:
        # Remove leading and trailing whitespaces (e.g., newline characters)
        name = line.strip()
        
        # Append the name to the list of names
        names.append(name)

# Print the total number of names and the longest name
print("Total number of names:", len(names))
print("Longest name:", max(names, key=len))

# Find and print the shortest names
shortest_names = find_shortest_name(names)
print("Shortest name(s):", shortest_names)
```

```python
import os
def write_sorted_names(names):
    # Sort the names in reverse alphabetical order
    sorted_names = sorted(names, reverse=True)

    # Open "sorted_data.txt" for writing (append mode if it exists)
    with open("sorted_data.txt", "a" if os.path.isfile("sorted_data.txt") else "w") as file:
        # Write the sorted names to the file, one name per line
        for name in sorted_names:
            file.write(name + "\n")

# Initialize an empty list to store names
names = []

# Open the file "data.txt" for reading
with open("data.txt", "r") as file:
    # Loop through each line in the file
    for line in file:
        # Remove leading and trailing whitespaces (e.g., newline characters)
        name = line.strip()

        # Append the name to the list of names
        names.append(name)

# Call the function to write the sorted names to "sorted_data.txt"
write_sorted_names(names)

print("Sorted names have been written to sorted_data.txt.")
```

```python
def search_for_name(name_to_search, names):
    # Convert both the name to search and the names in the list to lowercase for case-insensitive comparison
    name_to_search = name_to_search.lower()
    names_lower = [name.lower() for name in names]

    if name_to_search in names_lower:
        # If the name is found, get its position in the list
        position = names_lower.index(name_to_search)
        found_name = names[position]
        print(f"Name found at position {position + 1}: {found_name}")
    else:
        # If the name is not found, display a message indicating it's not in the list
        print(f"{name_to_search.capitalize()} is not in the list.")

# Initialize an empty list to store names
names = []

# Open the file "data.txt" for reading
with open("data.txt", "r") as file:
    # Loop through each line in the file
    for line in file:
        # Remove leading and trailing whitespaces (e.g., newline characters)
        name = line.strip()

        # Append the name to the list of names
        names.append(name)

# Get the name to search from the user
name_to_search = input("Enter a name to search for: ")

# Call the search_for_name function
search_for_name(name_to_search, names)
```

```python
import os

# Function to find the shortest names (case-insensitive)
def find_shortest_name(names):
    # Initialize variables to store the shortest length and a list of names with that length
    shortest_length = float('inf')
    shortest_names = []

    for name in names:
        name_length = len(name)
        if name_length < shortest_length:
            # Reset the list with the new shortest name
            shortest_names = [name]
            shortest_length = name_length
        elif name_length == shortest_length:
            # Append the name to the list of shortest names
            shortest_names.append(name)

    return shortest_names

# Function to filter names by starting letter
def filter_names_by_starting_letter(names, starting_letter):
    matching_names = [name for name in names if name.lower().startswith(starting_letter.lower())]
    return matching_names

# Function to write a new name to "names.txt" (append mode) with input validation
def write_new_name_to_file():
    while True:
        new_name = input("Enter a new name to add to the file (or press Enter to quit): ").strip()
        
        if new_name:
            with open("names.txt", "a") as file:
                file.write(new_name + "\n")
            print(f"'{new_name}' has been added to the 'names.txt' file.")
        else:
            print("Error: Name cannot be empty. Please enter a valid name.")
            continue
        
        add_more = input("Do you want to add another name? (y/n): ").lower()
        if add_more != 'y':
            break

# Initialize an empty list to store names
names = []

# Open the file "data.txt" for reading
with open("data.txt", "r") as file:
    # Loop through each line in the file
    for line in file:
        # Remove leading and trailing whitespaces (e.g., newline characters)
        name = line.strip()
        
        # Append the name to the list of names
        names.append(name)

# Display the shortest names (case-insensitive)
shortest_names = find_shortest_name(names)
print("Shortest names (case-insensitive):", shortest_names)

# Ask the user for a starting letter and filter names by that letter
starting_letter = input("Enter a starting letter to filter names (case-insensitive): ").strip()
filtered_names = filter_names_by_starting_letter(names, starting_letter)
if filtered_names:
    print(f"Names starting with '{starting_letter}' (case-insensitive):", filtered_names)
else:
    print(f"No names found starting with '{starting_letter}' (case-insensitive).")

# Write a new name to "names.txt" with input validation
write_new_name_to_file()
```

```python
import json
from datetime import datetime

# Step 1: Create an empty dictionary called student_records
student_records = {}

# Step 2: Implement a function add_student to add new student records
def add_student(regno, name, dob):
    student_records[regno] = {
        "Name": name,
        "DoB": dob
    }

# Step 3: Add three student records using the add_student function
add_student("2021001", "Alice", "2000-01-15")
add_student("2021002", "Bob", "2001-03-20")
add_student("2021003", "Charlie", "1999-07-05")

# Step 4: Write the student_records dictionary to a JSON file
with open("student_records.json", "w") as json_file:
    json.dump(student_records, json_file, indent=4)

# Step 5: Implement a function load_student_records to read the JSON file
def load_student_records(filename):
    with open(filename, "r") as json_file:
        return json.load(json_file)

# Step 6: Load data from "student_records.json" into loaded_student_records
loaded_student_records = load_student_records("student_records.json")

# Step 7: Print the contents of loaded_student_records
print("Loaded Student Records:")
print(json.dumps(loaded_student_records, indent=4))

# Step 8: Implement a function update_student to update student information
def update_student(regno, name, dob):
    if regno in student_records:
        student_records[regno]["Name"] = name
        student_records[regno]["DoB"] = dob
    else:
        print(f"Student with regno {regno} not found.")

# Step 9: Use the update_student function to update information
update_student("2021002", "Bob Smith", "2001-03-20")

# Step 10: Write the updated student_records dictionary to "student_records.json"
with open("student_records.json", "w") as json_file:
    json.dump(student_records, json_file, indent=4)

# Step 11: Load data from "student_records.json" into updated_student_records
updated_student_records = load_student_records("student_records.json")

# Step 12: Print the contents of updated_student_records
print("\nUpdated Student Records:")
print(json.dumps(updated_student_records, indent=4))

# Step 13: Implement a function calculate_age to calculate student age
def calculate_age(dob):
    dob_date = datetime.strptime(dob, "%Y-%m-%d")
    today = datetime.today()
    age = today.year - dob_date.year - ((today.month, today.day) < (dob_date.month, dob_date.day))
    return age

# Step 14: Calculate and add student ages to the records
for regno, student_info in student_records.items():
    student_info["Age"] = calculate_age(student_info["DoB"])

# Step 15: Write the updated student_records dictionary (with age) to "student_records.json"
with open("student_records.json", "w") as json_file:
    json.dump(student_records, json_file, indent=4)

# Step 16: Load data from "student_records.json" into final_student_records
final_student_records = load_student_records("student_records.json")

# Step 17: Print the contents of final_student_records
print("\nFinal Student Records (with Age):")
print(json.dumps(final_student_records, indent=4))

```

```python
import re
from collections import Counter

# Sample text data (you can replace this with your own string)
text_data = """
This is a sample text. It contains words with special characters, such as @, $, and %.
Some words are longer than others, and some words appear multiple times.
"""

# A. Calculate and display the total number of words in the text
words = text_data.split()
total_words = len(words)
print(f"A. Total number of words: {total_words}")

# B. Determine and display the longest word in the text along with its length
longest_word = max(words, key=len)
longest_word_length = len(longest_word)
print(f"B. Longest word: '{longest_word}' (Length: {longest_word_length})")

# C. Find and display the most frequently occurring word(s) in the text
word_counts = Counter(words)
most_common_words = word_counts.most_common(5)  # Change 5 to the desired number of results
print("C. Most frequently occurring word(s) and their frequencies:")
for word, frequency in most_common_words:
    print(f"'{word}': {frequency} times")

# D. Calculate and display the average word length in the text
total_word_length = sum(len(word) for word in words)
average_word_length = total_word_length / total_words
print(f"D. Average word length: {average_word_length:.2f} characters")

# E. Identify and display the word(s) that contain the most special characters
special_character_count = lambda word: len(re.findall(r'[^a-zA-Z0-9]', word))
words_with_most_special_chars = max(words, key=special_character_count)
most_special_chars_count = special_character_count(words_with_most_special_chars)
print(f"E. Word(s) with the most special characters:")
print(f"'{words_with_most_special_chars}' (Special characters: {most_special_chars_count})")
```
